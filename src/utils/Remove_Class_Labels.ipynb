```python
import os
```

```python
train_path = "./labels/train/"

for filename in os.listdir(train_path):
  if '.txt' in filename:
    with open(train_path+filename, "r") as f:
      lines = f.readlines()
    with open(train_path+filename, "w") as f:
        for line in lines:
          if line.startswith("0 "):
              f.write(line)
          elif line.startswith("1 "):
            line = list(line)
            line[0] = "0"
            line = "".join(line)
            f.write(line)
```

```python
val_path = "./labels/validation/"

for filename in os.listdir(val_path):
  if '.txt' in filename:
    with open(val_path+filename, "r") as f:
      lines = f.readlines()
    with open(val_path+filename, "w") as f:
        for line in lines:
          if line.startswith("0 "):
              f.write(line)
          elif line.startswith("1 "):
            line = list(line)
            line[0] = "0"
            line = "".join(line)
            f.write(line)
```
