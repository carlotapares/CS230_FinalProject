```python
import os
import cv2
import numpy as np
```

```python
train_path = "./train/"
val_path = "./validation/"
test_path = "./test/"
```

```python
img = cv2.imread('../images/train/050821_0813_OCHF_OP_0654.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
img_size = img.shape
```

```python
number_im = 0
number_otter = []
otter_sizes = []
wa = []
ha = []
xc = []
yc = []
for filename in os.listdir(train_path):
  if 'txt' in filename:
    number_im += 1
    number_otter.append(0)
    f = open(train_path + filename, 'r')
    for line in f:
      line = line.rstrip()
      data = line.split(' ')
      box = list(map(float,data[1:]))
      w = box[2] * img_size[1]
      h = box[3] * img_size[0]
      wa.append(w)
      ha.append(h)
      xc.append(box[0] * img_size[1])
      yc.append(box[1] * img_size[1])
      px_otter = w*h
      otter_sizes.append(px_otter)
      number_otter[-1] += 1
```

```python
print(number_im, np.sum(number_otter), np.count_nonzero(number_otter), np.mean(number_otter),np.mean(otter_sizes)/(img_size[0]*img_size[1]))
print(np.mean(wa)/img_size[1],np.mean(ha)/img_size[0],np.mean(xc)/img_size[1],np.mean(yc)/img_size[0])
```
