```python
import shutil, os
import random
```

```python
train_path = "./train/"
val_path = "./validation/"
test_path = "./test/"
```

```python
training_filenames = []

for filename in os.listdir(train_path):
  if '.jpeg' in filename:
    training_filenames.append(filename)
```

```python
val_filenames = []

for filename in os.listdir(val_path):
  if '.jpeg' in filename:
    val_filenames.append(filename)
```

```python
test_filenames = []

for filename in os.listdir(test_path):
  if '.jpeg' in filename:
    test_filenames.append(filename)
```

```python
print(len(training_filenames),len(val_filenames),len(test_filenames))
```

```python
train_size = 700
val_size = 75
test_size = 75
```

```python
missing_val_files = val_size - len(val_filenames)
missing_test_files = test_size - len(test_filenames)
```

```python
files_to_append = random.sample(range(0, len(training_filenames)), missing_val_files+missing_test_files)

val_files_to_append = files_to_append[:missing_val_files]
test_files_to_append = files_to_append[missing_val_files:]
```

```python
for f in val_files_to_append:
  filename = training_filenames[f]
  shutil.move(train_path + filename, val_path)
```

```python
for f in test_files_to_append:
  filename = training_filenames[f]
  shutil.move(train_path + filename, test_path)
```

```python
%cd /content/drive/MyDrive/SeaOtterDrone2021/sea_otter_dataset/
```

```python
images_path = "./images/"
labels_path = "./dataset_otters/labels/"
train_path = "train/"
val_path = "validation/"
test_path = "test/"
```

```python
im_training_filenames = []

for filename in os.listdir(images_path + train_path):
  if '.jpeg' in filename:
    im_training_filenames.append(filename)

lb_training_filenames = []

for filename in os.listdir(labels_path + train_path):
  if '.txt' in filename:
    lb_training_filenames.append(filename)
```

```python
im_val_filenames = []

for filename in os.listdir(images_path + val_path):
  if '.jpeg' in filename:
    im_val_filenames.append(filename)

lb_val_filenames = []

for filename in os.listdir(labels_path + val_path):
  if '.txt' in filename:
    lb_val_filenames.append(filename)
```

```python
im_test_filenames = []

for filename in os.listdir(images_path + test_path):
  if '.jpeg' in filename:
    im_test_filenames.append(filename)

lb_test_filenames = []

for filename in os.listdir(labels_path + test_path):
  if '.txt' in filename:
    lb_test_filenames.append(filename)
```

```python
print(len(im_training_filenames),len(im_val_filenames),len(im_test_filenames))
print(len(lb_training_filenames),len(lb_val_filenames),len(lb_test_filenames))
```

```python
for filename in im_val_filenames:
  filename = filename.split('.')[0] + '.txt'
  if filename not in lb_val_filenames:
    shutil.move(labels_path + train_path + filename, labels_path + val_path)
for filename in im_test_filenames:
  filename = filename.split('.')[0] + '.txt'
  if filename not in lb_test_filenames:
    if filename in lb_training_filenames:
      shutil.move(labels_path + train_path + filename, labels_path + test_path)
    elif filename in lb_val_filenames:
      shutil.move(labels_path + val_path + filename, labels_path + test_path)
    else:
      print('File not found', filename)
```

```python
for filename in im_training_filenames:
  filename = filename.split('.')[0] + '.txt'
  if filename not in lb_training_filenames:
    print(filename)
```

```python
train_size = 700
val_size = 70
test_size = 70
```

```python
missing_train_files = train_size - len(im_training_filenames)
over_val_files = len(im_val_filenames) - val_size
over_test_files = len(im_test_filenames) - test_size
```

```python
print(missing_train_files, over_val_files, over_test_files)
```

```python
files_to_append = random.sample(range(0, len(im_val_filenames)), over_val_files)

for f in files_to_append:
  filename = im_val_filenames[f].split('.')[0]
  shutil.move(labels_path + val_path + filename + '.txt', labels_path + train_path)
  shutil.move(images_path + val_path + filename + '.jpeg', images_path + train_path)
```

```python
files_to_append = random.sample(range(0, len(im_test_filenames)), over_test_files)

for f in files_to_append:
  filename = im_test_filenames[f].split('.')[0]
  shutil.move(labels_path + test_path + filename + '.txt', labels_path + train_path)
  shutil.move(images_path + test_path + filename + '.jpeg', images_path + train_path)
```
